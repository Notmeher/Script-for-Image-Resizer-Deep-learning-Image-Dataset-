RESIZE THE IMAGE SIZE

```python
import os
from PIL import Image

def resize_images(src_path, dest_path, new_size=(256, 256)):
    src_folder_list = os.listdir(src_path)
    
    for src in src_folder_list:
        folder_path = os.path.join(src_path, src)
        dest_folder_path = os.path.join(dest_path, src)
        
        # Create the destination folder if it doesn't exist
        os.makedirs(dest_folder_path, exist_ok=True)

        image_list = os.listdir(folder_path)
        
        for img in image_list:
            img_path = os.path.join(folder_path, img)
            
            try:
                # Check if the file is an image
                if any(img.lower().endswith(ext) for ext in ('.jpg', '.jpeg', '.png', '.gif', '.bmp')):
                    # Open the image using Pillow
                    with Image.open(img_path) as image:
                        # Resize the image
                        resized_image = image.resize(new_size)
                        
                        # Save the resized image with the same file name
                        resized_image.save(os.path.join(dest_folder_path, img))
                        print(f"Resized and saved {img} successfully.")
                else:
                    print(f"Skipping non-image file: {img}")
            except Exception as e:
                print(f"Error processing {img}: {e}")

# Example usage with a resize size of (128, 128)
src_path = "D:/Final code/Final dataset/PD/Train"
dest_path = "D:/Final code/Final dataset/PD/Train/resized_training"
resize_size = (64, 64)
resize_images(src_path, dest_path, new_size=resize_size)
```
